## Model Development:

### Importing libraries:

```python
import pandas as pd
import numpy as np
```

### Opening the file:

```python
df = pd.read_csv('cleaned_laptops.csv')
df.head()
```

## One Hot encoding:

```python
com_dummies = pd.get_dummies(df.Company)
com_dummies
```

```python
type_dummies = pd.get_dummies(df['TypeName'])
type_dummies
```

```python
os_dummies = pd.get_dummies(df['OpSys'])
os_dummies
```

```python
df = pd.concat([df,com_dummies,type_dummies,os_dummies],axis='columns')
df.head()
```

```python
df = df.drop('Company',axis='columns')
df = df.drop('TypeName',axis='columns')
df = df.drop('OpSys',axis='columns')
```

```python
df.head()
```

## Final dataset:

```python
df = df.drop(['Product','Inches','ScreenResolution','Cpu','Gpu','Memory','Weight_in_kgs'],axis='columns')
df.head()
```

## Building a Model

```python
X = df.drop(['Price_in_Rupees'],axis='columns')
X.head()
```

```python
y = df['Price_in_Rupees']
y.head()
```

### Splitting into Train & Test:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

### Linear Regression model:

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

### K Fold cross validation -- to check accuracy:

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

### Finding best model using GridSearchCV:

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

Since, Linear Regression is the best fit. We can use it (that we have already created)

### Testing the model:

```python
def predict_price(Ram_in_GB,Company,TypeName,OpSys):    
    loc_index = np.where(X.columns==Company)[0][0]
    loc_index1 = np.where(X.columns==TypeName)[0][0]
    loc_index2 = np.where(X.columns==OpSys)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = Ram_in_GB

    if loc_index >= 0:
        x[loc_index] = 1
    if loc_index1 >= 0:
        x[loc_index1] = 1
    if loc_index2 >= 0:
        x[loc_index2] = 1
    
    print('Price: ',lr_clf.predict([x])[0]) 
```

```python
predict_price('2','Lenovo','Notebook','Windows 10')
```
